```r
a = grid_pred(c("SOI", "BIENSO"),-1,"mean")
b = grid_pred(c("pr"),-1,"sum")
c = grid_pred(c("STORAGE"),seq(12),"mean")

predictors = c(a,b,c)
predictors
```

```r
#input data from preprocess
data_input <- preprocess_data(datetime_initialisation = '2022-11-01',
                              forecast_mode = "cv",
                              catchment_code = "4703002",
                              predictor_list = predictors)
```

```r
source("base/Regression_model.R")
library('caret')
X_train = data_input$X_train
y_train = data_input$y_train$volume
resampling_method = 'LOOCV'
preProcess = c("center", "scale")
method = 'lm'
trcontrol = trainControl(method = resampling_method,savePredictions = "all")
X_train  %>% names 
```

```r
# remove correlated predictors
cor_matrix <- cor(X_train)
high_cor <- findCorrelation(cor_matrix, cutoff = 0.8, verbose = FALSE)
X_train_uncorr <- X_train[, -high_cor]
X_train_uncorr  %>% names
```

```r
library(ggplot2)
ggplot(data = log(data_input$y_train))+geom_histogram(aes(x = volume),bins = 5)
```

```r
regression_model = 
train(
  X_train_uncorr,
  y_train,
  metric = "RMSE",
  trControl = trcontrol,
  method = method,
  preProcess = preProcess
)
```

```r
# determine variable importance
index_imp = which(varImp(regression_model)$importance > 0.1)
# remove predictors with low importance
cols = colnames(X_train_uncorr)

x_reduced <- data.frame(X_train_uncorr[,index_imp])
colnames(x_reduced) = cols[index_imp]
x_reduced  %>% names
```

```r
# train model with reduced predictors
model_reduced <- train(x_reduced,
                       y_train,
                       method = method,
                       preProcess = preProcess,
                       trControl = trcontrol,
                       tuneGrid = regression_model$bestTune)
```

```r
# compare performance of full and reduced models
merge(regression_model$bestTune,regression_model$results)$RMSE
model_reduced$results$RMSE
```
